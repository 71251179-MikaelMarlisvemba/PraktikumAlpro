Latihan 7.1

```python
def prima(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

def terdekat(m):
    for i in range(m-1, 1, -1):
        if prima(i):
            return i
    return None

m = int(input("Masukkan Bilangan : "))
dekat = terdekat(m)
if dekat:
    print(f"Bilangan Prima Terdekat dari {m} adalah {dekat}")
else:
    print("Tidak Ada Lagi Bilangan Prima Dibawah", m)
```

Latihan 7.2

```python
n = int(input("Masukkan nilai n: "))

for i in range(n, 0, -1):
    hasil = 1
    for j in range(1, i + 1):
        hasil *= j
    print(hasil, end=" ")

    for k in range(i, 0, -1):
        print(k, end=" ")
    print()  
```

```python
tinggi = int(input("Masukkan Tinggi Deret : "))
lebar = int(input("Masukkan Lebar Deret : "))

mulai = 1
selesai = lebar

for i in range(1, tinggi + 1):
    for j in range(mulai, selesai + 1):
        print(j, end = " ")
    selesai += lebar
    mulai += lebar
    print()
```

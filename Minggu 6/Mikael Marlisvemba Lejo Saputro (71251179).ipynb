```python
def perkalian(a, b):
    i = 0
    if a == 1 or b == 1 or b == -1 or a == -1:
        print(f"{a * b}")
    elif a == 0 or b == 0:
        print(0)
    elif a < 0:
        print("-( ", end = '')
        while a != i:
            print(f"{b}", end = ' ')
            if i > a + 1:
                print('+', end=' ')
            i -= 1
        print(f") = {a * b}")
    elif b < 0:
        while a != i:
            print(f"({b})", end = ' ')
            if i < a - 1:
                print('+', end=' ')
            i += 1
        print(f"= {a * b}")
    else:
        while a != i:
            print(b, end = ' ')
            if i < a - 1:
                print('+', end=' ')
            i += 1
        print(f"= {a * b}")

a = int(input("Masukkan Nilai a : "))
b = int(input("Masukkan Nilai b : "))
if a < 0 and b < 0:
    print(f"({a}) x ({b}) =", end = ' ')
elif b < 0:
    print(f"{a} x ({b}) =", end = ' ')
elif a < 0:
    print(f"({a}) x {b} =", end = ' ')
else:
    print(f"{a} x {b} =", end = ' ')
perkalian(a, b)
```

Latihan 6.2

```python
def ganjil(bawah, atas):
    if bawah < atas:
        for i in range(bawah, atas + 1):
            if i % 2 != 0:
                print(i, end='')
                if i < atas - 1:
                    print(end = ', ')
    else: 
        for i in range(bawah, atas - 1, -1):
            if i % 2 != 0:
                print(i, end='')
                if i > atas + 1:
                    print(end = ', ')
                
bawah = int(input("Input Bawah : "))
atas = int(input("Input Atas : "))
ganjil(bawah, atas)    
```

```python
m = int(input("Berapa Jumlah Mata Kuliah : "))
total_nilai = 0

for i in range(1, m + 1):
    huruf = input(f"Nilai MK {i} : ")
    if huruf == "A":
        nilai = 4
    elif huruf == "B":
        nilai = 3
    elif huruf == "C":
        nilai = 2
    elif huruf == "D":
        nilai = 1
    else:
        print("Nilai Tidak Valid")
        nilai = 0
        break
    total_nilai += nilai
IPS = total_nilai/m
print(f"Nilai IPS anda semester ini {IPS:.2f}")
```

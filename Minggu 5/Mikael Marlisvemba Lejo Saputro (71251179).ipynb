Latihan 5.1

```python
def cek_angka(a,b,c):
    if (a != b and b != c) and ((a + b == c) or (a + c == b) or (b + c == a)):
        return True
    else:
        return False
    
a = int(input("Input a : "))
b = int(input("Input b : "))
c = int(input("Input c : "))
print(f"Nilai : {cek_angka(a,b,c)}")
```

Latihan 5.2

```python
def cek_digit_belakang(a,b,c):
    a = a % 10
    b = b % 10
    c = c % 10

    if a == b or b == c or a == c:
        return True
    else:
        return False

a = int(input("Nilai a : "))
b = int(input("Nilai b : "))
c = int(input("Nilai c : "))
print(f"Nilai : {cek_digit_belakang(a,b,c)}")
```

```python
def konversi_suhu(c, ke):
    if ke == "F":
        return (f"Celcius to Fahrenheit. F = {(9/5) * c + 32: .0f}.")
    elif ke == "R":
        return (f"Celcius to Reamur. R = {0.8 * c: .0f}.")
    else:
        return ("Input Tidak Valid")


c = int(input("Input Suhu : "))
ke = input("Konversi Ke : ")

print(konversi_suhu(c, ke))
```

```python
inputsuhu = input("Masukkan Suhu Tubuh: ")

try:
    suhu = int(inputsuhu)
    if suhu >= 38:
        print("Anda demam")
    else:
        print("Anda tidak demam")
except:
    print("Input Harus Berupa Angka")
```

```python
inpbilangan = input("Masukkan suatu bilangan: ")

try:
    bilangan = int(inpbilangan)
    if bilangan > 0:
        print("Positif")
    elif bilangan < 0:
        print("Negatif")
    elif bilangan == 0:
        print("Nol")
except:
    print("Input Harus Berupa Angka")
```

```python
a = input("Masukkan bilangan pertama: ")
b = input("Masukkan bilangan kedua: ")
c = input("Masukkan bilangan ketiga: ")

try:
    a = int(a)
    b = int(b)
    c = int(c)
    if a > b and a > c:
        print("Terbesar: ", a)
    elif b > a and b > c:
        print("Terbesar: ", b)
    elif c > a and c > b:
        print("Terbesar: ", c)
except:
    print("Input Harus Berupa Angka")
```

```python
inputbln = input("Masukkan Bulan (1 - 12) : ")

try :
    bulan = int(inputbln) 
    if 0 < bulan <= 12:
        if bulan == 1 or bulan == 3 or bulan == 5 or bulan == 7 or bulan == 8 or bulan == 10 or bulan == 12:
            print("Jumlah Hari : 31")
        elif bulan == 4 or bulan == 6 or bulan == 9 or bulan == 11:
            print("Jumlah Hari : 30")
        elif bulan == 2:
            print("Jumlah Hari : 29")
    else :
        print("Bulan Tidak Valid")
except:
    print("Input Bulan Tidak Valid")
```

```python
a = input("Masukkan sisi 1: ")
b = input("Masukkan sisi 2: ")
c = input("Masukkan sisi 3: ")

try:
    a = int(a)
    b = int(b)
    c = int(c)

    if a == b == c:
        print("3 Sisi Sama")
    elif (a == b and b != c) or (a != b and a == c) or (b == c and a != c):
        print("2 Sisi Sama")
    else:
        print("Tidak Ada Sisi Yang Sama")
except:
    print("Input Harus Berupa Angka")
```

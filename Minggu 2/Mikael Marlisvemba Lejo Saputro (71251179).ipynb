Latihan 2.1

```python
BMI = float(input("Masukkan Body Mass Index yang Anda Harapkan : "))
cm = int(input("Masukkan Tinggi Badan Anda dalam cm: "))
tinggi = cm / 100

berat = BMI * tinggi**2
print(f"BMI : {BMI} ")
print(f"Tinggi Badan Anda : {tinggi} m")
print(f"Berat Badan Yang Dibutuhkan Untuk Mencapai BMI yang Anda Harapkan adalah : {berat :0.2f} Kg")
```

```python
x = int(input("Masukkan Nilai x : "))
f = (2 * x**3) + (2 * x) + (15 / x)

print(f"Hasil dari fungsi f({x}) = {f:.2f}")
```

Latihan 2.3

```python
gaji_per_jam = int(input("Gaji Per Jam yang Diinginkan : Rp. "))
jam_kerja = int(input("Jumlah Jam Kerja Per Minggu : "))

print("")
print(f"Gaji Yang Diinginkan : Rp. {gaji_per_jam :,}/jam")
print(f"Jam Kerja Per Minggu : {jam_kerja} jam/minggu")
print("Bekerja Selama 5 Minggu")
print("")

gaji = (gaji_per_jam * jam_kerja) * 5
pendapatan_sebelum = gaji

pajak = gaji * 0.14
pendapatan_sesudah = pendapatan_sebelum - pajak

pakaian_aksesoris = pendapatan_sesudah * 0.1
alat_tulis = pendapatan_sesudah * 0.01
sisa_uang = pendapatan_sesudah - (pakaian_aksesoris + alat_tulis)

sedekah = sisa_uang * 0.25
anak_yatim = sedekah * 0.3
dhuafa = sedekah - anak_yatim
sisa = sisa_uang - sedekah

print(f"1. Pendapatan Budi Sebelum Membayar Pajak : Rp. {pendapatan_sebelum :,.0f}")
print(f"2. Pendapatan Budi Sesudah Membayar Pajak : Rp. {pendapatan_sesudah :,.0f}")
print(f"3. Jumlah uang yang akan Budi habiskan untuk membeli pakaian dan aksesoris : Rp. {pakaian_aksesoris :,.0f}")
print(f"4. Jumlah uang yang akan Budi habiskan untuk membeli alat tulis : Rp. {alat_tulis :,.0f}")
print(f"Sisa Uang Setelah Membeli baju, aksesoris dan alat tulis : Rp. {sisa_uang :,.0f}")
print(f"5. Jumlah uang yang akan Budi sedekahkan : Rp. {sedekah :,.0f}")
print(f"6. Jumlah uang yang akan diterima anak yatim : Rp. {anak_yatim :,.0f}")
print(f"7. Jumlah uang yang akan diterima kaum dhuafa : Rp. {dhuafa :,.0f}")
print(f"Sisa Uang Setelah Sedekah : Rp. {sisa :,.0f}")
```
